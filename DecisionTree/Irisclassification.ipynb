```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LinearRegression

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
iris
```

```python
iris_df = pd.DataFrame(iris.data,columns=iris.feature_names)
iris_df["Species"]=iris.target
iris_df.head()
```

```python
iris.keys()
```

```python
iris.target_names
```

```python
# X and Y values are
X=iris_df.drop(["Species"],axis=1)
y=iris_df["Species"]

# Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=17)
print("The diamension of X_train is{}".format(X_train.shape))
print("The diamension of X_test is{}".format(X_test.shape))
```

```python
dt_classifier=DecisionTreeClassifier()
dt_classifier.fit(X_train,y_train)
```

```python
dt_classifier=DecisionTreeClassifier()
dt_classifier.fit(X_train,y_train)

score=dt_classifier.score(X_train,y_train)
print("R Squared Train : ",score)

score=dt_classifier.score(X_test,y_test)
print("R Squared Test : ",score)

y_pred=dt_classifier.predict(X_test)

mse=mean_squared_error(y_test,y_pred)
print("MSE:",mse)
print("RMSE:",mse**(1/2.0))

```

```python
from sklearn import tree
#from sklearn import plottree
plt.figure(figsize=(15,10))
tree.plot_tree(dt_classifier,filled=True)
```

```python
dt_classifier=DecisionTreeClassifier(criterion="entropy",max_depth=3)
dt_classifier.fit(X_train,y_train)

score=dt_classifier.score(X_train,y_train)
print("R Squared Train : ",score)

score=dt_classifier.score(X_test,y_test)
print("R Squared Test : ",score)

y_pred=dt_classifier.predict(X_test)

mse=mean_squared_error(y_test,y_pred)
print("MSE:",mse)
print("RMSE:",mse**(1/2.0))

```

```python
lr=LinearRegression()
lr.fit(X_train,y_train)

score=lr.score(X_train,y_train)
print("R Squared Train : ",score)

score=lr.score(X_test,y_test)
print("R Squared Test : ",score)

y_pred=lr.predict(X_test)

mse=mean_squared_error(y_test,y_pred)
print("MSE:",mse)
print("RMSE:",mse**(1/2.0))
```
